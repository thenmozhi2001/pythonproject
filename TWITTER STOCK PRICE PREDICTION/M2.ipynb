# Data Visualisation

```python
import warnings
warnings.filterwarnings('ignore')
```

### Importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('twitter-stocks.csv')
```

### Let's See Heatmap of the dataset

```python
# Heatmap plot diagram
fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(data.corr(), ax=ax, annot=True)
print('These parameters are more related ')
```

### Pair Plot

```python
sns.pairplot(data)
plt.show()
```

### High Price Distribution.

```python
fig, ax = plt.subplots(figsize=(15,6))
sns.boxplot(data.High, ax =ax)
plt.title("High price")
plt.show()
```

### Daily deviations of high price

```python

plt.plot(data['Date'],data['High'])
plt.xlabel('Date')
plt.ylabel('High Price')
plt.title('Daily Deviation Of High Price')
plt.show()
```

### This is the relation between High Price and open price of every day

```python
plt.scatter(data['High'],data['Open'])
plt.show()
```

### Relation between High price and closed Price of every day.

```python
plt.scatter(data['High'],data['Close'])
plt.show()
```
